### 1. Load Data


```python
import sys
import subprocess
import warnings
warnings.filterwarnings("ignore")

try:
    import mlxtend
except ImportError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "mlxtend"])

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx
import difflib

sns.set_theme(style='whitegrid')
```

```python
basket = pd.read_csv("../Data/basket.csv",index_col =0)
```

```python
basket.head()
```

### 2. Apply FP-Growth


```python
frequent_items = fpgrowth(
    basket[basket.sum(axis=1) >= 2],
    min_support=0.0001,
    use_colnames=True,
    max_len=2
)

frequent_items = frequent_items.sort_values(by="support", ascending=False)

frequent_items.head(20)
```

### 3. Frequent Product Pairs

```python
frequent_pairs = frequent_items[frequent_items["itemsets"].apply(lambda x: len(x) >= 2)]

frequent_pairs.head(20)
```

### 4. Generate Association Rules


```python
rules = association_rules(
    frequent_items,
    metric="lift",
    min_threshold=1
)

rules = rules.sort_values(by="lift", ascending=False)

rules.head(10)
```

### 5. Clean Final Rules Table

```python
rules_display = rules.copy()

rules_display["antecedents"] = rules_display["antecedents"].apply(lambda x: ", ".join(list(x)))
rules_display["consequents"] = rules_display["consequents"].apply(lambda x: ", ".join(list(x)))

final_rules = rules_display[["antecedents", "consequents", "support", "confidence", "lift"]]

final_rules.head(20)
final_rules.to_csv("../Data/Final_rules.csv")
```

### 5.1 Advanced Visualizations

```python
# Top Rules - 3-panel chart
top_rules = rules.head(10).copy()
top_rules['rule'] = top_rules['antecedents'].apply(lambda x: ', '.join(list(x))) + ' → ' + top_rules['consequents'].apply(lambda x: ', '.join(list(x)))

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Top 10 Association Rules - Key Metrics', fontsize=14, fontweight='bold')

metrics = ['confidence', 'support', 'lift']
colors = ['steelblue', 'mediumseagreen', 'coral']

for ax, metric, color in zip(axes, metrics, colors):
    sr = top_rules.sort_values(metric)
    bars = ax.barh(sr['rule'], sr[metric], color=color, edgecolor='white', alpha=0.9)
    ax.set_title(metric.capitalize(), fontsize=12, fontweight='bold')
    ax.set_xlabel(metric)
    ax.tick_params(axis='y', labelsize=8)
    for bar, v in zip(bars, sr[metric].values):
        ax.text(bar.get_width()+0.001, bar.get_y()+bar.get_height()/2,
                f'{v:.4f}', va='center', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Metric Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(rules[['support', 'confidence', 'lift']].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.3f', linewidths=0.5)
plt.title('Correlation Matrix of Rule Metrics', fontsize=14, fontweight='bold')
plt.show()
```

### 6. Recommendation Function

```python
def recommend_products(product_name, rules_table=final_rules, top_n=5, threshold=0.6):
    # Get unique antecedents
    unique_products = rules_table['antecedents'].unique().tolist()
    
    # Fuzzy matching to find the closest product name
    matches = difflib.get_close_matches(product_name, unique_products, n=1, cutoff=threshold)
    
    if not matches:
        print(f"Product '{product_name}' not found. No close matches available.")
        return pd.DataFrame()
        
    matched_product = matches[0]
    if matched_product != product_name:
        print(f"Product not found exactly. Using closest match: '{matched_product}'")
    else:
        print(f"Found exact match: '{matched_product}'")
        
    recommendations = rules_table[rules_table["antecedents"] == matched_product]
    recommendations = recommendations.sort_values(by="lift", ascending=False)
    
    return recommendations[["antecedents", "consequents", "confidence", "lift"]].head(top_n)

# Example usage with a misspelled product name
example_misspelled = "Mouse Chartruse 292"
print(f"Input: {example_misspelled}")
recommend_products(example_misspelled)
```
